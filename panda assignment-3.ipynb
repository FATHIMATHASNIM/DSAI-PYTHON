```python
import pandas as pd
import numpy as np
```

```python
#1Get the Metadata from the above files.
df1=pd.read_csv(r'https://raw.githubusercontent.com/jackiekazil/data-wrangling/master/data/chp3/data-text.csv')

print(df1.info())
```

```python
df2=pd.read_csv(r'https://raw.githubusercontent.com/kjam/data-wrangling-pycon/master/data/berlin_weather_oldest.csv')
df2
print(df2.info())
```

```python
#2. Get the row names from the above files.
df1.head()
print('row names of dataset 1')
for row in df1.index: 
    print(row, end = " \t")
```

```python
df2.head()
print("row names of dataset 2")
for row in df2.index: 
    print(row, end = " \t")
```

```python
#3. Change the column name from any of the above file and store the changes made permanently.
df1.rename(columns={'PUBLISH STATES':'PUBLISH STATUS'},inplace=True)
df1
```

```python
#4. Change the names of multiple columns.
df1.rename(columns={'Country':'region','Indicator':'INDICATOR','Year':'YEAR','WHO region':'WHO REGION','Sex':'GENDER','Comments':'MESSAGE'},inplace=True)
df1
```

```python
#5. Arrange values of a particular column in ascending order.
df1.sort_values('Display Value',ascending=True)
```

```python
#6. Arrange multiple column values in ascending order.
df2.sort_values(['TMAX','TMIN'],ascending=True)
```
